Conditionals (if,elif,else)
Conditionals let a program make decisions.
"If something id true --> do this, otherwise --> that"

Basic syntax:
if condition:
    runs if condition is true
elif another_condition:
    runs if above is false and this is true
else:
    runs if all conditions are false


```python
#simple if statement
age=20
if age>=18:
    print("you can vote")   #if condition agrees then only it will print
if age>21:
    print("You can vote")   #here conditions are not matching so not printed
```

```python
#if-else statement

rain=True                          #if we take condition true than take umbrella will be printed and 
                                   #if we take condition false then don't take umbrella will be printed
if rain:
    print("Take Umbrella")
else:
    print("don't take umbrella")
```

```python
#if-else

marks=35
if marks>=40:
    print("pass")
else:
    print("fail")
```

```python
#elif statement

weather="sunny"

if weather=="rain":
    print("Take umbrella")
elif weather=="sunny":
    print("Take sunglasses")
else:
    print("don't take umbrella")
```

```python
#if-elif-else

score=78
if score>=90:
    print("Grade A")
elif score>=75:
    print("Grade B")
elif score>=60:
    print("Grade C")
else:
    print("Grade D")
```

```python
#checking temperature(example)

#use if else and print it is hot if temperature is >25 if not print not hot

temperature=25
if temperature>25:
    print("it is hot")
else:
    print("not hot")
```

```python
#even or odd

num=8
if num%2==0:
    print("It is even")
else:
    print("It is odd")
```

```python
#login check

username="admin"
password="1234"

if username=="admin" and password=="1234":
    print("login successfull")
else:
    print("invalid login credentials")
if username=="admin" or password=="1234":
    print("loggedin")
```

structural pattern matching(match-case)

match-case is pythons's switch statement

use it when:
you compare one variable against many fixed values

Basic syntax:

match value:
    case option1:
        #code
    case option2:
        #code
    case option3:
        #code
    case_:
        #default

here _ means "anything else"

```python
#menu selection

choice=int(input("Enter your choice"))


match choice:
    case 1:
        print("Add item")
    case 2:
        print("Delete item")
    case 3:
        print("Update item")
    case _:
        print("invalid choice")
```

```python
#using string

day="Sunday"

match day:
    case "Saturday" | "Sunday":               # or is used as "|"
        print("Weekend")
    case _:
        print("Weekday")
```

```python
#simple calculator option

operation="+"

match operation:
    case "+":
        print("Addition selected")                           #match -> check this value
    case "-":                                                #case  -> if it matches this
        print("Subtraction selected")                        # -    -> anything else
    case "*":
        print("Multiplication selected")
    case "/":
        print("Division selected")
    case _:
        print("invalid operation")
```

match is not for conditions like greater than or less than
match cannot be used for a range of values either

case age>18 #invalid
case num>30 #invalid

```python
#Truthy or Falsy

#What should happen if a list or variable is empty

items=[]

if items:
    print("items exist")
else:
    print("no items")

#python treats empty things as flase
```
