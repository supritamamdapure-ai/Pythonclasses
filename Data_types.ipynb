```python
#padding
num=42
print(f"{num:06d}") #05d is number to expand before the number #padding
```

```python
#alignment
text="Python"
print(f"{text:<10}")  #left alignment (10 is no. of characters)
print(f"{text:^10}")  #center alignment
print(f"{text:>10}")  #right alignment
```

```python
new='this is a string'
print(new[6])  #positive indexing
print(new[-5])   #negative indexing (counts from backwards)
```

```python
#slicing strings (dividing your strings into multiple parts)
#string[start:stop:step]
apple='this is a new string'
print(apple[1:6])               #it will not consider last number
print(apple[:])      #entire text
print(apple[:7])     #stop at 7th letter
print(apple[::2])    #every 2nd element
print(apple[-5:-1])
```
